#Gradio Interface

```python
translator = Translator()

def translate_logic(text):
    if not text.strip():
        return "Please enter some text."


    try:
        detection = translator.detect(text)
        detected_lang = detection.lang


        if detected_lang == 'en':
            target = 'bn'
        elif detected_lang == 'bn':
            target = 'en'
        else:
            return f"Unsupported language detected ({detected_lang}). Please use English or Bangla."


        result = translator.translate(text, dest=target)
        return result.text

    except Exception as e:
        return f"Error: {str(e)}"





demo = gr.Interface(
    fn=translate_logic,
    inputs=gr.Textbox(label="Enter English or Bangla text", placeholder="Type here..."),
    outputs=gr.Textbox(label="Translation"),
    title="English ↔ Bangla Auto-Translator",
    description="Type in English to get Bangla, or type in Bangla to get English.",
    allow_flagging="never"
)

# Launch the interface
demo.launch()
```
